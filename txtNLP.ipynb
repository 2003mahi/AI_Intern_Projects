```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')

def split_data_by_sentiment(data):
    """
    Splits a list of text data into two lists based on sentiment polarity.

    Args:
        data: A list of strings, where each string is a piece of text.

    Returns:
        A tuple containing two lists:
            - positive_data: A list of text data with positive sentiment.
            - negative_data: A list of text data with negative sentiment.
    """

    analyzer = SentimentIntensityAnalyzer()
    positive_data = []
    negative_data = []
    for text in data:
        scores = analyzer.polarity_scores(text)
        if scores['compound'] >= 0.05:  # Adjust threshold as needed
            positive_data.append(text)
        elif scores['compound'] <= -0.05:
            negative_data.append(text)
    return positive_data, negative_data

# Example usage:
data = [
    "This is a great movie!",
    "I hate this product.",
    "The weather is beautiful today.",
    "I'm feeling very sad.",
    "The food was delicious.",
    "This is a terrible experience."
]

positive_texts, negative_texts = split_data_by_sentiment(data)

print("Positive Texts:")
for text in positive_texts:
    print(text)

print("\nNegative Texts:")
for text in negative_texts:
    print(text) # Added indentation here to include this line in the loop
```
